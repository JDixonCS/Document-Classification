```python
import pandas as pd
from datasets import Dataset

# Load CSV file as a pandas DataFrame
df = pd.read_csv("526892.csv", encoding="cp1252", index_col=0)
print(df)

# Convert DataFrame to dictionary
data_dict = {"sentence": list(df["sentence"]), "label": list(df["label"])}
print(data_dict)

# Convert dictionary to Dataset object
dataset = Dataset.from_dict(data_dict)
print(dataset)
```
